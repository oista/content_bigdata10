```python
df = pd.read_csv("../datasets/DP_LIVE_04042018210609495.csv", sep=",")
```

```python
rus_df = df[df.LOCATION == "RUS"]
```

```python
plt.title("GDP of Russian Federation per capita")
plt.ylabel("USD")
plt.plot(rus_df.TIME, rus_df.Value, color='green', marker='.', linestyle='solid')
plt.show()
```

```python
plt.boxplot(rus_df.Value.values, vert=False)
plt.xlim(xmin=0)
plt.grid()
plt.show()
```

```python
df_2016 = df[df.TIME==2016]
plt.pie(df_2016.Value.values, labels=df_2016.LOCATION, autopct='%1.1f%%', shadow=True)
plt.show()
```

```python
df_2016 = df[df.TIME==2016]
sample_2016 = df_2016.sample(10)

plt.pie(sample_2016.Value.values, labels=sample_2016.LOCATION, autopct='%1.1f%%', shadow=True)
plt.axis('equal')
plt.show()
```

```python
plt.figure(figsize=(25,8))
plt.rc('ytick', labelsize=10)
plt.rc('xtick', labelsize=15)
plt.grid()

countries_nums = range(df_2016.shape[0])
plt.bar(countries_nums, df_2016.Value.values)
plt.xticks(countries_nums, df_2016.LOCATION.values, rotation='vertical')
plt.show()
```
